```python
import cv2
import numpy as np
"""drawing contours starts here"""
#step 1: load the image 
image = cv2.imread('images/bunchofshapes.jpg')
cv2.imshow("original image",image)
cv2.waitKey(0)

#step 2: draw a blank image 
blank_image = np.ones(image.shape)*0
cv2.imshow("blank image",blank_image)
cv2.waitKey(0)

#step 3:create a copy of original image 
copy_image = image
copy_image_2 = image
cv2.imshow("copy image",copy_image)
cv2.waitKey(0)

#step 4:create a gray image 
gray_image = cv2.cvtColor(copy_image,cv2.COLOR_BGR2GRAY)
cv2.imshow("gray image",gray_image)
cv2.waitKey(0)

#step 5:canny edge detection
edge_image = cv2.Canny(gray_image,100,200)
cv2.imshow("edge image",edge_image)
cv2.waitKey(0)

#step 6:find contours
contours , hierarchy = cv2.findContours(edge_image,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE) 

#step 7:now draw the contours
cv2.drawContours(blank_image,contours,-1,255,3)
cv2.imshow("only contours",blank_image)
cv2.waitKey(0)
cv2.imwrite("only contours.jpg",blank_image)
cv2.drawContours(copy_image,contours,-1,255,3)
cv2.imshow("contours on image",copy_image)
cv2.imwrite("contours on image.jpg",copy_image)

"""drawing contours part completed here"""

"""sorting contours part starts here"""
#step 8:
def area_Of_Contours(contours):
    area_list = []
    for contour in contours:
        area = cv2.contourArea(contour)
        area_list.append(area)
    return area_list
print "area of contours without sorting ",area_Of_Contours(contours)

#step9: sorting the contours 
sorted_contours = sorted(contours, key=cv2.contourArea,reverse =True)
print "area of contours with sorting ",area_Of_Contours(sorted_contours)

#step 10: Iterate over our contours and draw one at a time
cv2.imshow("sorting contours of image",copy_image_2)
for i in range(len(sorted_contours)) :
    cv2.drawContours(copy_image_2,sorted_contours,i,[0,255,0],3)
    cv2.imshow("sorted contours by area",copy_image_2)
    cv2.imwrite("sorted contours by area"+str(i+1)+".jpg",copy_image_2)
    cv2.waitKey(0)

#step 12:sorting left to right or vis versa

def x_cord_contour(contours):
    #Returns the X cordinate for the contour centroid
    if cv2.contourArea(contours) > 10:
        M = cv2.moments(contours)
        return (int(M['m10']/M['m00']))

    
def center_of_contour(image, c):
    # Places a red circle on the centers of contours
    M = cv2.moments(c)
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])
 
    # Draw the countour number on the image
    cv2.circle(image,(cx,cy), 10, (0,0,255), -1)
    return image


copy_image_3 = image.copy()

#step 13: draw the center of mass on image
for c in contours:
    center = center_of_contour(copy_image_3,c)
    
cv2.imshow("image with center of mass",copy_image_3)
cv2.imwrite("image with center of mass.jpg",copy_image_3)

#step 14:sorting left 2 right
sort_contour_L2R = sorted(contours,key =x_cord_contour,reverse=False )

for (i,c) in enumerate(sort_contour_L2R):
    M = cv2.moments(c)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    cv2.putText(image, str(i+1), (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("sorted image right2left",image)
    cv2.imwrite("sorted image right2left.jpg",image)
    cv2.waitKey(0)
    
    #cropping
    (x, y, h, w) = cv2.boundingRect(c) 
    
    cropped_contour = image[y:y+w,x:x+h]
    image_name = "output_shape_number_" + str(i+1) + ".jpg"
    cv2.imshow(image_name,cropped_contour)
    cv2.imwrite(image_name,cropped_contour)

#step 11:close all windows 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
