```python
import cv2 
import numpy as np

def face_check(image):
    my_face = cv2.imread("C:\Users\Atul Kumar Singh\Pictures\Camera Roll\WIN_20170922_093020.JPG")
    my_face_gray = cv2.cvtColor(my_face, cv2.COLOR_BGR2GRAY)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    sift = cv2.SIFT()
    
    key1, des1 = sift.detectAndCompute(my_face_gray,None)
    key2, des2 = sift.detectAndCompute(gray,None)
    
    bf = cv2.BFMatcher()
    
    matches = bf.knnMatch(des1,des2,k=2)
    good_matches = []
    for m,n in matches:
        if m.distance < 0.7 * n.distance:
            good_matches.append(m) 

    return len(good_matches)

cap= cv2.VideoCapture(0)

while True :
    
    ret , frame = cap.read()
    #gray_1 = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    frame_copy = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    find_face = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
    face = find_face.detectMultiScale(frame_copy, 1.3, 5)
    
    match = face_check(frame)
    x_ = 0
    y_ = 0
    for (x,y,x2,y2) in face:
        x_ = x
        y_ = y
        cv2.rectangle(frame,(x,y),(x+x2,y+y2),(0,255,0),2)
        
    thresold = 50
    print match
    if match > thresold :
        
        cv2.putText(frame,"face_recogonized",(x_,y_-5),2,1,(0,255,0),2)
    else :
        cv2.putText(frame,"face_not_recogonized",(x_,y_-5),2,1,(0,255,0),2)
        
    cv2.imshow('face recognizer', frame)
    cv2.imwrite('face recognizer.jpg', frame)
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
    
cap.release()
cv2.destroyAllWindows()  
    
    
    
```
