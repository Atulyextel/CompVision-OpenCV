```python
import cv2
import numpy as np

image = cv2.imread('images/input.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

surf = cv2.SURF()

surf.hessianThreshold  = 2000

key , des = surf.detectAndCompute(gray,None)
print "key = ",key
print "des = ",des

image = cv2.drawKeypoints(image,key, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2.imshow("surf",image)
cv2.imwrite("surf.jpg",image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
